Example 1 plots 'cosine' and 'sine' functions in the range [0 $2\pi$) usign subplots sharing x-axis and example 2 plots 'Signum' and 'Exponential' functions in the range [-1 1) usign subplots sharing y-axis.

```python
# Example 1
import numpy as np 
import matplotlib.pyplot as pyplot

x=np.arange(0, 2*np.pi, np.pi/10)
#print(x)
y=np.cos(x)
z=np.sin(x)
f,axarr = pyplot.subplots(2, sharex=True)   #Sharing X-axis
axarr[0].stem(x,y)
axarr[0].set_title('Cosine plot')
axarr[1].stem(x,z)
axarr[1].set_title('Sine plot')
#print(z)
#pyplot.stem(x,y);ig
pyplot.show()
```

```python
# Example 2
import numpy as np 
import matplotlib.pyplot as pyplot 
x=np.arange(-1, 1, 0.1)
y=np.sign(x)
z=np.exp(x)
f, (ax1, ax2) = pyplot.subplots(1, 2, sharey=True)  #Sharing Y axis
ax1.stem(x,y)
ax1.set_title('Signum function')
ax2.stem(x,z)
ax2.set_title('Exponential function')

pyplot.show()
```

```python
# Example 3
import numpy as np 
import matplotlib.pyplot as pyplot 

x=np.arange(0, 2*np.pi, np.pi/15)
y1=2*np.cos(x)
y2=np.cos(x)
y3=0.5*np.cos(x)
fig, ax = pyplot.subplots()

#Dash style dashes = [20, 5, 20, 5]  means 20 points on, 5 off, 20 on, 5 off
line1 = ax.plot(x, y1, '--', linewidth=2, dashes=[20, 5, 20, 5], label='2cos(x)')

#Line styles
line2 = ax.plot(x,y2, linestyle='--', color='r', linewidth=3, label='cos(x)')

#Marker style
line3 = ax.plot(x,y3, marker='v', linewidth=1, label='0.5cos(x)')

ax.legend(loc='lower right')

pyplot.show()
```

## Log plot

```python
# log y axis
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.01, 20.0, 0.01)
plt.semilogy(t, np.exp(-t/5.0))
plt.title('semilogy')
plt.grid(True)
plt.show()
```

```python
# log x axis
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.01, 20.0, 0.01)
plt.semilogx(t, np.sin(2*np.pi*t))
plt.title('semilogx')
plt.grid(True)
plt.show()
```

```python
# log x and y axis
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
t = np.arange(0.01, 20.0, 0.01)
plt.loglog(t, 20*np.exp(-t/10.0), basex=2)
plt.grid(True)
plt.title('loglog base 2 on x')
plt.show()
```

## Plotting Histogram

The NumPy histogram function applied to an array returns a pair of vectors: the histogram of the array and the vector of bins. Beware: matplotlib also has a function to build histograms (called hist, as in Matlab) that differs from the one in NumPy. The main difference is that pylab.hist plots the histogram automatically, while numpy.histogram only generates the data.

```python
import numpy as np
import matplotlib.pyplot as plt
# Build a vector of 10000 normal deviates with variance 0.5^2 and mean 2
mu, sigma = 2, 0.5
v = np.random.normal(mu,sigma,10000)
# Plot a normalized histogram with 50 bins
plt.hist(v, bins=50, normed=1)       # matplotlib version (plot)
plt.show()
```

```python
# Compute the histogram with numpy and then plot it
(n, bins) = np.histogram(v, bins=50, normed=True)  # NumPy version (no plot)
plt.plot(.5*(bins[1:]+bins[:-1]), n)
plt.show()

```

## Plotting histogram for an image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img=mpimg.imread('cameraman.png',0)
imgplot = plt.imshow(img)
plt.show()

(n, bins) = np.histogram(img, bins=255)  # NumPy version (no plot)
plt.plot(.5*(bins[1:]+bins[:-1]), n)
plt.show()
```
